```python
from pymongo import MongoClient
import pandas as pd
from pprint import pprint
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the uk_food database to a variable name
db = mongo['uk_food']
```

```python
# Review the collections in our database
db.list_collection_names()
```

```python
# assign the collection to a variable
establishments = db['establishments']
```

```python
# Find the establishments with a hygiene score of 20
query = {'scores.Hygiene': {'$eq': 20}}
result = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print(establishments.count_documents({'scores.Hygiene': {'$eq': 20}}))

# Check if result is not empty before trying to access the first document
if result:
    from pprint import pprint
    pprint(result[0])  # Display the first document in the results
else:
    print("No establishments found with a hygiene score of 20.")
```

```python
# Convert the result to a Pandas DataFrame
hygiene_20_df = pd.DataFrame(list(establishments.find(query)))
# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(hygiene_20_df)}")
# Display the first 10 rows of the DataFrame
hygiene_20_df.head(10)
```

```python
# Find the establishments with London in the LocalAuthorityName and has a RatingValue greater than or equal to 4.
query = {'LocalAuthorityName':{'$regex': 'London'},
        'RatingValue':{'$gte':'4'}}

results = list(establishments.find(query))

# Use count_documents to display the number of documents in the result
print("Number of documents in result:", len(results))

# Display the first document in the results using pprint
pprint(results[0])
```

```python
# Convert the result to a Pandas DataFrame
london_rating_df = pd.DataFrame(results)

# Display the number of rows in the DataFrame
print(f"Number of rows in DataFrame: {len(london_rating_df)}")

# Display the first 10 rows of the DataFrame
london_rating_df.head(10)
```

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match': {'scores.Hygiene': 0}}
group_query = {'$group': {'_id': '$LocalAuthorityName', 
                          'count': {'$sum': 1}
                         }
              }
sort_query = {'$sort': {'count': -1}}

pipeline = [match_query, group_query, sort_query]

results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print(f"There are {len(results)} documents in the result.")

# Print the first 10 results
pprint(results[0:10])
```
